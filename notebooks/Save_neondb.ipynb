```python
from dotenv import load_dotenv
import os
load_dotenv()
from pyspark.sql import SparkSession
spark = SparkSession.builder \
    .appName("MyApp") \
    .config("spark.jars", "/usr/local/spark/jars/postgresql-42.2.18.jar") \
    .getOrCreate()
df = spark.read.parquet("./dados/gov/bolsa_familia")

USER = os.getenv('PGUSER')
PASSWORD = os.getenv('PGPASSWORD')
```

```python
neon_url = "jdbc:postgresql://ep-billowing-math-ahww7eyr-pooler.c-3.us-east-1.aws.neon.tech/minha_base"
```

```python
df.write \
    .format("jdbc") \
    .option("url", neon_url) \
    .option("dbtable", "bolsa_familia") \
    .option("user", USER) \
    .option("password", PASSWORD) \
    .option("driver", "org.postgresql.Driver") \
    .save()
```
